```python
# Dependencies
import pandas as pd
```

```python
# ask user to enter directory path for gene file 
gene_file_path_input = input("enter path of working directory: ") # ask user ot enter gene file path
```

```python
# open and read gene file
gene_file = open(gene_file_path_input, "r") 
gene_file = gene_file.readlines()
gene_file
```

```python
# clean gene file and convert to panda DataFrame
gene_clean_data = [i.replace (",\t", ",").strip('\n').split('\t') for i in gene_file]
col_names = ['gene', 'variant', 'chr_num', 'gene_start', 'gene_end','strand', 'gene_location']
df = pd.DataFrame(gene_clean_data, columns = col_names)
df
```

```python
# find uniqe genes
unique_genes = df['gene'].unique()
unique_genes
```

```python
# find variant per gene
gene_variant_count = pd.DataFrame(df.groupby("gene")["variant"].count())
gene_variant_count
```

```python
# find genes with single variant
single_variant_genes= gene_variant_count[(gene_variant_count == 1).any(axis=1)]
single_variant_genes
```

```python
# find genes with multiple variants
multiple_variants_genes= gene_variant_count[(gene_variant_count != 1).any(axis=1)]
multiple_variants_genes
```

```python
# genes per ‘+’ and ‘-’ strand
strand_gene_count = pd.DataFrame(df.groupby("strand")["gene"].count())
strand_gene_count
```

```python
# find genes per chromosome
chromosome_gene_count = pd.DataFrame(df.groupby("chr_num")["gene"].count())
chromosome_gene_count
```

```python
# find variant per chromosome
chromosome_variant_count = pd.DataFrame(df.groupby("chr_num")["variant"].count())
chromosome_variant_count
```
